2. Створіть функцію, що реалізує критерій Шапіро-Уілка та застосуйте її до кожної змінної набору данних. 

```python

def shapito(DataKeyFrame,iSrednee,normB):
    
    sum2=0
    for i in range(20):
        sum2+=(DataKeyFrame[i]-iSrednee)**2
    tempB=0
    for i in range(10):
        tempB+=normB[i]*(DataKeyFrame[20-i-1]-DataKeyFrame[i])
    B=(tempB**2)/sum2
    return B
    
```
