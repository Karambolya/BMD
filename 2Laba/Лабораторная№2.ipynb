```python
import numpy as np
```

```python
x=np.linspace(0,5,15)
y=x**2
x
```

```python
y
```

# Завдання 2.1

```python
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel("NotTime")
plt.title("Functional Method")
```

```python
plt.subplot(1,2,1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("X Y")
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.xlabel('Y')
plt.ylabel('X')
plt.title("Y x")
plt.plot(y,x,'b')
```

# Завдання 2.2

```python
figure=plt.figure()
axes=figure.add_axes([1,1,1,1])
axes.plot(x,y)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Name123")
```

```python
figur=plt.figure()
axes1=figur.add_axes([0,0,1,1])
axes2=figur.add_axes([0.1,0.5,0.3,0.3])
axes1.plot(x,y,'p')
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("Name123")

axes2.plot(y,x,'x')
axes2.set_xlabel("Y")
axes2.set_ylabel("x")
axes2.set_title("Name1234")
```

```python
figure,axes=plt.subplots(nrows=1,ncols=1)
axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title("Name")
```

```python
figure,axes=plt.subplots(nrows=2,ncols=1)
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'x')
```

```python
figure,axes=plt.subplots(nrows=5,ncols=5)
```

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend()
```

```python

fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,x**2,color='r')
axes[0][1].plot(x,x**2,color='b')
axes[1][0].plot(x,x**2,lw=2,ls='--')
axes[1][1].plot(x,x**2,marker='o',markersize=10,markerfacecolor='r')
```
