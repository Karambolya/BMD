```python
import seaborn as sns

import numpy as np
#%matplotlib inline
```

```python
diamonds=sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
sns.distplot(diamonds['carat'])
```

```python
sns.distplot(diamonds['carat'],kde=False)
```

```python
sns.jointplot(x='carat',y='depth',data=diamonds)
```

```python
sns.jointplot(x='carat',y='depth',data=diamonds,kind='hex')
```

```python
sns.pairplot(diamonds.head())
```

```python
sns.pairplot(diamonds.head(),hue="cut")
```

```python
sns.barplot(x='cut',y="price",data=diamonds,estimator=np.median)
```

```python
sns.boxplot(x='color',y="price",data=diamonds,palette="Set2")
```

```python
sns.stripplot(x='color',y='table',data=diamonds)
```
